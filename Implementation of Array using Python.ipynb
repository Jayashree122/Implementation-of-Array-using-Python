```python
#Implementation of Array
class Array:
  def __init__(self,size):
    self.a=[None]*size
    self.size=size
    self.n=0
  def isfull(self):
    return self.size==self.n
  def isempty(self):
    return self.n==0
  def len_(self):
    return self.n
  def ins_at_beg(self,element):
    if self.isfull():
      print("Array is full")
    else:
      self.n+=1
      for i in range(self.n-1,1,-1):
        self.a[i]=self.a[i-1]
      self.a[0]=element
  def ins_at_end(self,element):
    if self.isfull():
      print("Array is full")
    else:
      self.n+=1
      self.a[self.n-1]=element
  def del_at_end(self):
    if self.isempty():
      print("Array is empty")
    else:
      self.n-=1
  def del_at_beg(self):
    if self.isempty():
      print("Array is empty")
    else:
      self.n-=1
      for i in range(self.n):
        self.a[i]=self.a[i+1]
  def search(self,target):
    if self.isempty():
      print("Array is empty")
    else:
      for i in range(self.n):
        if self.a[i]==target:
          return "Present-"
      return -1
  def sort_(self):
    if self.isempty():
      print("Array is empty")
    else:
      for i in range(self.n):
        for j in range(i+1,self.n):
          if self.a[i]>self.a[j]:
            self.a[i],self.a[j]=self.a[j],self.a[i]
  def print_array(self):
    if self.isempty():
      print("Array is empty")
    else:
      for i in range(self.n):
        print(self.a[i])
  def reverse_(self):
    if self.isempty():
      print("Array is empty")
    else:
      i=0
      j=self.n-1
      while i<j:
        self.a[i],self.a[j]=self.a[j],self.a[i]
        i+=1
        j-=1
  def ins_at_position(self,element,position):
    if self.isfull():
      print("Array is full")
    else:
      self.n+=1
      for i in range(self.n-1,position,-1):
        self.a[i]=self.a[i-1]
        self.a[position]=element
  def del_at_position(self,position):
    if self.isempty():
      print("Array is empty")
    else:
      self.n-=1
      for i in range(self.n,position):
        self.a[i]=self.a[i+1]

```

```python
obj=Array(5)
obj.isfull()
obj.isempty()
obj.ins_at_beg(5)
obj.ins_at_end(7)
obj.ins_at_position(3,1)
obj.ins_at_position(6,2)
obj.sort_()
obj.reverse_()
obj.print_array()
obj.search(7)
obj.del_at_beg()
obj.del_at_end()
obj.del_at_position(2)
obj.print_array()
```
